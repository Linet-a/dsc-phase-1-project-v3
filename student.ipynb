## Data Cleaning

```python
#check unique values in the investingation type column
df['Investigation.Type'].unique()

#identify rows with valid types 
valid_type = ['Accident', 'Incident']
invalid_rows = df[~df['Investigation.Type'].isin(valid_type)]
invalid_rows


# replace the invalid values with 'unknown'

df['Investigation.Type'] = df['Investigation.Type'].where(df['Investigation.Type'].isin(valid_type), 'unknown')
df['Investigation.Type'].unique() # checking if the values are replaced
```

```python
#filling the country column null values with the mode

df['Country'] = df['Country'].fillna(df['Country'].mode()[0])
```

```python
#filling missing  location with 'unknown' data

df['Location'] = df['Location'].fillna('Unknown Location')
```

```python
 # drop 'Injury.Severity' column as it contains several unique values 
df = df.drop(columns=['Injury.Severity'])
```

```python
#fiil Aircraft.Category FAR.Description, Schedule with mode values 
df['Aircraft.Category'] = df['Aircraft.Category'].fillna(df['Aircraft.Category'].mode()[0])
df['FAR.Description'] = df['FAR.Description'].fillna(df['FAR.Description'].mode()[0])
df['Schedule'] = df['Schedule'].fillna(df['Schedule'].mode()[0])
df['Aircraft.damage'] = df['Aircraft.damage'].fillna(df['Aircraft.damage'].mode()[0])
```

```python
df['Broad.phase.of.flight'] = df['Broad.phase.of.flight'].fillna('Uknown')
```

```python
#drop longitude and latitude colums and Airport code

df = df.drop(columns=['Latitude'])
df = df.drop(columns=['Longitude'])
df = df.drop(columns=['Airport.Code'])
df = df.drop(columns=['Airport.Name'])
```

```python
#fill Country miss with mode values
df[ 'Country'] = df['Country'].fillna(df['Country'].mode()[0])
```

```python
#replace missing null dnumerical value with the median

df['Total.Fatal.Injuries'] = df['Total.Fatal.Injuries'].fillna(df['Total.Fatal.Injuries'].median())
df['Total.Serious.Injuries'] = df['Total.Serious.Injuries'].fillna(df['Total.Serious.Injuries'].median())
df['Total.Minor.Injuries'] = df['Total.Minor.Injuries'].fillna(df['Total.Minor.Injuries'].median())
df['Total.Uninjured'] = df['Total.Uninjured'].fillna(df['Total.Uninjured'].median())
```

```python
#Replace crafts with null registration number with 'Unknown'
df['Registration.Number'] = df['Registration.Number'].fillna('Uknown')
df['Number.of.Engines'] = df['Number.of.Engines'].fillna('Uknown')
```

```python
#replace the small missing data with the most frequent
df['Make'] = df['Make'].fillna(df['Make'].mode()[0])
df['Model'] = df['Model'].fillna(df['Model'].mode()[0])
df['Amateur.Built'] = df['Amateur.Built'].fillna(df['Amateur.Built'].mode()[0])
df['Engine.Type'] = df['Engine.Type'].fillna(df['Engine.Type'].mode()[0])
df['Purpose.of.flight'] = df['Purpose.of.flight'].fillna(df['Purpose.of.flight'].mode()[0])
df['Weather.Condition'] = df['Weather.Condition'].fillna(df['Weather.Condition'].mode()[0])
```

```python
df['Report.Status'].head(10) #inspecting information on 'Report.Status'
```

```python
#keeping report status for cross-analysis and droping it from the dataframe
report_status_copy = df['Report.Status']

df = df.drop(columns=['Report.Status'])
```

```python
df['Publication.Date'].isna().sum()

#replace the missing values with'Not Indicated'
df['Publication.Date'] = df['Publication.Date'].fillna('Not Indicated')
```

# Data  Visualization

## Distribution of Accident numbers by Aircraft Type

```python
# Prepare data for the bar chart
aircraft_category_counts = df['Aircraft.Category'].value_counts()

# Data for grouping insignificant values into "Other" for the pie chart
threshold = 0.05  # Categories contributing less than 5% will be grouped into 'Other'
total_count = aircraft_category_counts.sum()
percentages = aircraft_category_counts / total_count
other_categories = percentages[percentages < threshold].index

# Group these into 'Other'
grouped_counts = aircraft_category_counts.copy()
grouped_counts.loc[other_categories] = 0  # Zero out small categories
grouped_counts['Other'] = aircraft_category_counts.loc[other_categories].sum()  
grouped_counts = grouped_counts[grouped_counts > 0] 

fig, axes = plt.subplots(1, 2, figsize=(14, 6))


axes[0].bar(aircraft_category_counts.index, aircraft_category_counts.values, color='#023e8a')
axes[0].set_title('Aircraft Category - Bar Chart', fontsize=14)
axes[0].set_xlabel('Aircraft Category')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Plot the pie chart in the second subplot
axes[1].pie(
    grouped_counts, 
    labels=grouped_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['#732E00', '#ffcc99']
)
axes[1].set_title('Aircraft Category Distribution (Grouped)', fontsize=14)

# Display the subplots
plt.tight_layout()
plt.show()
```

## Number of accidents Recorded per Engine Type

```python
plt.figure(figsize=(10,6))


accidents_per_engine_type = df['Engine.Type'].value_counts().sort_values(ascending=False).plot(kind='bar', color='#023e8a')


plt.title('Number of Accidents per Engine Type' , fontsize=14)
plt.xlabel('Engine Type', fontsize=14)
plt.ylabel('Number of Accidents', fontsize=20)
plt.xticks(rotation=45)
plt.legend(fontsize=20)
plt.tight_layout()
plt.show()
```

## Aircraft Make and Models by Accident Count

```python

# Group by 'Make' and 'Model' and count the number of accidents
accidents_per_make_model = df.groupby(['Make', 'Model']).size()

# top 10 and accidents
top_accidents = accidents_per_make_model.nlargest(10).sort_values(ascending=True)

plt.figure(figsize=(15,6))
top_accidents.plot(kind='barh', color='Skyblue')


plt.title('Top 10 Accidents per Make and Model', fontsize=14)
plt.xlabel('Number of Accidents', fontsize=12)
plt.ylabel('Make and Model', fontsize=12)
plt.tight_layout()


plt.show()
```

## Additional Analysis

### Time Plot Series of Accidents Over Time

```python
# converting the date strings into date.time onjects
df['Event.Date'] = pd.to_datetime(df['Event.Date'])

# Grouping the data into dates
accidents_per_day = df.groupby(df['Event.Date'].dt.date).size()


plt.figure(figsize=(12, 6))
plt.plot(accidents_per_day.index, accidents_per_day.values, color='#023e8a')

plt.title('Number of Accidents Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=45)


plt.tight_layout()
plt.show()
```

## Accidents Recorded Per Location Vs their Weather Common Patterns

```python
#getting locations with the most common accident occurence
accident_top_locations = df['Location'].value_counts().head(10)
weather_in_top_locations = df[df['Location'].isin(accident_top_locations.index)].groupby(['Location', 'Weather.Condition']).size().unstack()
plt.figure(figsize=(12, 6))
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=False)

#Most common accident sites plot
accident_top_locations.plot(kind='bar', ax=axes[0], color='#732E00')
axes[0].set_title('Most Common Accident Sites', fontsize=14)
axes[0].set_xlabel('Location', fontsize=12)
axes[0].set_ylabel('Number of Accidents', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

#Weather patterns in those locations
weather_in_top_locations.plot(kind='bar', stacked=True, ax=axes[1], color=['#023e8a','#732E00', '#ffcc99' ])
axes[1].set_title('Weather Patterns in Top Accident Sites', fontsize=14)
axes[1].set_xlabel('Location', fontsize=12)
axes[1].set_ylabel('Weather Condition Frequency', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)
```

## Most Onboarded  Aircraft

```python
# Calculate the total onboard count for each entry
df['Total.Onboard'] = (df['Total.Fatal.Injuries'] + 
                       df['Total.Serious.Injuries'] + 
                       df['Total.Minor.Injuries'] + 
                       df['Total.Uninjured'])

# total onboarding per make per model
most_onboarded_aircraft = df.groupby(['Make', 'Model'])['Total.Onboard'].sum().sort_values(ascending=False)

# top 10 most onboarded aircraft
most_onboarded_aircraft.head(10)


most_onboarded_aircraft.head(10).plot(kind='bar', figsize=(10, 6), color='#204D00')
plt.title('Top 10 Most Onboarded Aircraft', fontsize=16)
plt.xlabel('Aircraft Make', fontsize=12)
plt.ylabel('Total Onboard Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Most Onboarded Aircraft with Least Accidents

```python

# Create a grouped DataFrame with total onboard and accident count
onboard_accidents = df.groupby('Make').agg(
    Total_Onboard=('Total.Onboard', 'sum'),
    Accident_Count=('Event.Id', 'count')
).reset_index()

# Filter for most onboarded with least accidents
onboard_least_accidents = onboard_accidents.sort_values(
    by=['Accident_Count', 'Total_Onboard'], ascending=[True, False]
).head(10)




fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(onboard_least_accidents['Make'], onboard_least_accidents['Total_Onboard'], color='#732E00', label='Total Onboard')
ax.set_xlabel('Total Onboard', fontsize=12)
ax.set_ylabel('Aircraft Make', fontsize=12)
ax.set_title('Most Onboarded Aircraft with Least Accidents', fontsize=16)
plt.gca().invert_yaxis()  # Invert Y-axis for better readability
plt.tight_layout()
plt.legend()
plt.show()
```
